```python
import numpy as np;
import pandas as pan;
import matplotlib.pyplot as plotter;
import seaborn as sns;
import utils as u
```

```python
matrix_data = pan.read_csv("Clean_Housing_Data.csv")
matrix_data.drop("Unnamed: 0", axis = 1, inplace=True)
```

```python
matrix_data["MS SubClass"] = np.vectorize(lambda val: str(val))(matrix_data["MS SubClass"])
```

```python
matrix_data.select_dtypes("object")
```

```python
numeric_only = matrix_data.select_dtypes(exclude= "object")
cat_only = matrix_data.select_dtypes("object")
```

```python
cat_only_dummies = pan.get_dummies(data = cat_only, drop_first = True)
```

```python
cat_only_dummies
```

```python
complete_matrix_data = pan.concat([cat_only_dummies, numeric_only], axis = 1)
```

```python
#we're on track, apparently
#so we can output the data as a csv file 
complete_matrix_data.to_csv("OneHotEncodedCleanData.csv", index = False)
```
